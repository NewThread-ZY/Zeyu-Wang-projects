### 2.4 Clean data

#### 2.4.1 Select which columns have Nan data

```python
data.count()
```

```python
(len(data)-data.count())/len(data)
```

We can see that the missing data of children and country are less than 0.05. The missing data of agent and company is greater than 0.05. So that we can remove the missing data of children and country and fix the data of agent and company.

```python
data1 = data.copy()
```

```python
data1 = data1[data1.children.notna()]#remove data which nan in children column
```

```python
data1 = data1[data1.country.notna()]#remove data which nan in country column
```

```python
data1
```

```python
data1.agent.describe()
```

In order to avoid overfitting in the latter train, I will use data that has the largest frequency to fill Nan data of agent.

```python
data1.agent.value_counts().head(1)
```

```python
data1.agent = data1.agent.fillna(9.0)
```

```python
data1.company.describe()
```

Although there are much missing data in the company column, we can't remove the company column because it has a large variance. So I need to fill it.

```python
data1.company.value_counts().head(1)
```

Consider the company data actually is category data rather than numerical data, I can't use mean or median numbers to fill it. I just can use the number which has the largest frequency to fill it.

```python
data1.company = data1.company.fillna(40)
```

```python
data1 = data1.reset_index(drop=True)
```

```python
data1
```

#### 2.4.3 Scalling data

I will use the max_min scalling method to scall data.

```python
data1.lead_time = (data1.lead_time-data1.lead_time.min())/(data1.lead_time.max()-data1.lead_time.min())
```

```python
data1.arrival_date_year = (data1.arrival_date_year-data1.arrival_date_year.min())#For year, we just need to minus the smallest year.
```

```python
data1.arrival_date_week_number = (data1.arrival_date_week_number-data1.arrival_date_week_number.min())#For week number, we just need to minus the smallest week number.
```

```python
data1.adr = (data1.adr-data1.adr.min())/(data1.adr.max()-data1.adr.min())
```

```python
data1.days_in_waiting_list = (data1.days_in_waiting_list-data1.days_in_waiting_list.min())/(data1.days_in_waiting_list.max()-data1.days_in_waiting_list.min())
```
