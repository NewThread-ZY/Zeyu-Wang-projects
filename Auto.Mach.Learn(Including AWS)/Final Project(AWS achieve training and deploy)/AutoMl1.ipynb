# 1. Import data

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/NewThread-ZY/AIM-5014-100/main/heart.csv')
data
```

```python
data.Sex = data.Sex.replace({'M': 0, 'F': 1})
data.ChestPainType = data.ChestPainType.replace({'ATA': 0, 'NAP': 1,'ASY': 2,'TA': 3})
data.RestingECG = data.RestingECG.replace({'Normal': 0, 'ST': 1,'LVH': 2})
data.ExerciseAngina = data.ExerciseAngina.replace({'N': 0, 'Y': 1})
data.ST_Slope = data.ST_Slope.replace({'Up': 0, 'Flat': 1, 'Down': 2})
data
```

```python
data = (data-data.min())/(data.max()-data.min())
data
```

# 2. Split data into trainning data and testing data

```python
import numpy as np
```

```python
train_data, test_data, _ = np.split(data.sample(frac = 1, random_state = 42), [int(0.8*len(data)), int(len(data))])
```

```python
#save to csv
train_data.to_csv('./heart_nor/AutoML_train.csv', index = False, header = True)
test_data.to_csv('./heart_nor/AutoML_test.csv', index = False, header = True)
```

```python
ep_name = 'Heart-AutoML1'
```

```python
import boto3,sys
sm_rt = boto3.Session().client('runtime.sagemaker')
```

```python
tp=tn=fp=fn=count=0
with open('./heart/AutoML_test.csv') as f:
    lines = f.readlines()
    for l in lines[1:]:
        l = l.split(',')
        label = l[-1]
        l = l[:-1]
        l = ','.join(l)
        response = sm_rt.invoke_endpoint(EndpointName=ep_name,ContentType='text/csv',Accept='text/csv',Body=l)
        response = response['Body'].read().decode('utf-8')
        if '1' in label:
            if '1' in response:
                tp+=1
            else:
                fn+=1
        if '0' in label:
            if '0' in response:
                tn+=1
            else:
                fp+=1
        count+=1
        if (count%100==0):
            sys.stdout.write(str(count)+' ')
print('done')
```

```python
print("%d %d" %(tp, fn))
print("%d %d" %(fp, tn))
```

```python
accuracy = (tp+tn)/(tp+tn+fn+fp)
accuracy
```

```python
precision = tp/(tp+fp)
precision
```

```python
recall = tn/(tn+fn)
recall
```

```python
f1 = (2*precision*recall)/(precision+recall)
f1
```
