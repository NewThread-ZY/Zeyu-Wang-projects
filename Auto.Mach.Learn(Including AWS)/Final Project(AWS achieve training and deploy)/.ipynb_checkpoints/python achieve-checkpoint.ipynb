```python
def clean_Latitude_Longitude(data):
    last_letter = data[-1]
    if last_letter=='N' or last_letter=='E':
        return float(data[:-1])
    else:
        return float('-'+data[:-1])
```

```python
df_GlobalLandTemperatures1_deletenan.Latitude = df_GlobalLandTemperatures1_deletenan.Latitude.apply(lambda x:clean_Latitude_Longitude(x))
df_GlobalLandTemperatures1_deletenan.Longitude = df_GlobalLandTemperatures1_deletenan.Longitude.apply(lambda x:clean_Latitude_Longitude(x))
```

d. Nonparameter regression

```python
def estimate(x, y, x0, h):
    w = np.exp( -(x-x0)**2 / h )
    W = np.diag(w)
    x = x[:,None]
    X = np.concatenate(
        (np.ones(shape=x.shape),
        (x-x0)), axis=1
    )
    B = np.linalg.inv(X.T @ W @ X) @ (X.T @ W @ y)
    return B[0]
```

```python
xhat = X_text_Latitude
yhat = np.array([estimate(np.array(X_train_Latitude), np.array(y_train), x0, 0.1) for x0 in xhat])
```

```python
plt.scatter(xhat, y_test)
plt.scatter(xhat, yhat, color='r')
```
