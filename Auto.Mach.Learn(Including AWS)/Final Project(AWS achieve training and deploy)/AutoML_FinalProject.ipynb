# 1. Import data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://gitlab.com/zwang9/aim-5014-100-finalproject/-/raw/master/GlobalLandTemperaturesByCity.csv')
data
```

# 2. Clean Data

```python
data0 = data.dropna()
```

```python
data1 = data0
```

```python
data1.dt = pd.to_datetime(data1.dt)
data1
```

```python
data1 = data1.set_index('dt')
```

```python
data1 = data1.loc['2000-01-01':'2012-12-31']
data1
```

```python
data1.isnull().sum()
```

```python
data1.groupby('City').count().describe()
```

```python
repeat_city = (data1.groupby(['City', 'Country']).count()[data1.groupby(['City', 'Country']).count().Longitude == 468]).index
repeat_city
```

```python
def clean_Latitude_Longitude(data):
    last_letter = data[-1]
    if last_letter=='N' or last_letter=='E':
        return float(data[:-1])
    else:
        return float('-'+data[:-1])
```

```python
data1.Latitude = data1.Latitude.apply(lambda x:clean_Latitude_Longitude(x))
data1.Longitude = data1.Longitude.apply(lambda x:clean_Latitude_Longitude(x))
```

```python
data1
```

```python
df_glt_gyll = data1.groupby(['Latitude', 'Longitude']).AverageTemperature.mean()
df_glt_gyll
```

```python
Latitude_Longitude = df_glt_gyll.index
```

# Question1

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Latitude_Longitude, df_glt_gyll, test_size=0.30, random_state=42)
```

```python
X_train = np.array(X_train)
X_train = np.array([list(i) for i in X_train])
X_test = np.array(X_test)
X_test = np.array([list(i) for i in X_test])
```

```python
X_train_Latitude = X_train[:,0]
X_train_Longitude = X_train[:,1]
X_test_Latitude = X_test[:,0]
X_text_Longitude = X_test[:,1]
```

```python
df_train_tem = pd.DataFrame({'Latitude':X_train_Latitude, 'Longitude':X_train_Longitude, 'y':y_train.values})
df_test_tem = pd.DataFrame({'Latitude':X_test_Latitude, 'Longitude':X_text_Longitude, 'y':y_test.values})
```

```python
df_train_tem.to_csv('./temperature/temperature_train.csv', index = False, header = True)
df_test_tem.to_csv('./temperature/temperature_test.csv', index = False, header = True)
```

#predict

```python
ep_name = 'temperature1'
```

```python
import boto3,sys
sm_rt = boto3.Session().client('runtime.sagemaker')
```

```python
tp=tn=fp=fn=count=0
error = 0
count = 0
with open('./temperature/temperature_test.csv') as f:
    lines = f.readlines()
    for l in lines[1:]:
        l = l.split(',')
        label = l[-1]
        l = l[:-1]
        l = ','.join(l)
        response = sm_rt.invoke_endpoint(EndpointName=ep_name,ContentType='text/csv',Accept='text/csv',Body=l)
        response = response['Body'].read().decode('utf-8')
        error+= (float(response)-float(label))**2
        count+=1
print('done')
```

```python
(error/count)**0.5
```
