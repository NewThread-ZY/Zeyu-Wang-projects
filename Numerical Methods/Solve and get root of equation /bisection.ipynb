```python
import numpy as np
```

```python
def bisection(f, a, b, tof):
    print(a,'++++++++++',b)
    aa = a
    bb = b
    if np.sign(f(a)) * np.sign(f(b)) > 0:
        print('f(a)f(b)>0 not satified')
        return
    n = 1
    fa = f(a)
    fb = f(b)
    while np.abs(a-b)>tof:
        c = (a+b)/2
        n = n+1
        fc = f(c)
        if np.isclose(fc,0):
            print('approximate root',c, 'has been obtained in', n, 'steps')
            '''if np.abs(aa-bb)==1:
                print(aa,' ',bb)'''
            return
        if np.sign(fa)*np.sign(fb)<0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc
    c = (a+b)/2
    '''if np.abs(aa-bb)==1:
        print(aa,' ',bb)'''
    print('the final interval [', a, b,'] contains a root')
    print('approximate root', c, 'has been obtained in', n, 'steps')
```

```python
def f(x):
    return (50*x**4+np.sin(x)-np.cos(x)+np.log(x)+x**(1/2))/x**2
```

Use Bisection method to find the solution.

```python
bisection(f, 0.00000000000000000001, 100, 0.000001)
```
