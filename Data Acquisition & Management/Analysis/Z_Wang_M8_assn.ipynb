# Part 3: Inferences

1. What proportion of diamonds are between .30 and 1.08 carats?(I will pick up the diamonds which its carat greater than 0.3 and I will get new pandas data, then I will pick up the diamonds which its carat greater than 1.8 again from the new data. Finally, use '/' to get the result for the question)

```python
carat_3 = data[data['carat']>0.3]
carat_3_8 = carat_3[carat_3['carat']<1.08]
proportion_3_8 = carat_3_8['carat'].count()/data['carat'].count()
print("Proportion of diamonds are between .30 and 1.08 carats is "+str(proportion_3_8))
```

2. How many of the diamonds have equal x and y dimensions?(I will pick up the data which their x is equal to y and count their Quantity)

```python
n_xy = data[data['x']==data['y']]['x'].count()
print("There are "+str(n_xy)+" diamonds have equal x and y dimensions")
```

3. How many of the diamonds have a carat value that is less than the mean carat value?(I will pick up the diamonds whose carat is less than the mean value and count them)

```python
n_caratlessmean = data[data['carat']<data['carat'].mean()]['carat'].count()
print("There are "+str(n_caratlessmean)+" diamonds have a carat value that is less than the mean carat value")
```

4. How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal.(Because better than Premium is only one, so I pick up them from data respectively and count them together)

```python
n_cutispandi = data[data['cut'] == 'Premium']['cut'].count()+data[data['cut'] == 'Ideal']['cut'].count()
print("There are "+str(n_cutispandi)+" diamonds have a Premium cut or better")
```

5. Which diamond has the highest price per carat? What is its value?(I will use the column price to divide by carat and use the new data column to create the new column. Finally, I can use the max() function to get the result)
