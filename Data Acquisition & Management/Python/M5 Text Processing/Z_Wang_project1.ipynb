```python
text = "555-1239Dr. Anthony Fauci(636) 555-0113Hollingdorp, Donnatella555-6542Fitzgerald, F. Scott555 8904Rev. Martin Luther King636-555-3226Snodgrass, Theodore5553642Carlamina Scarfoni"
```

### 1. (6 Points) Extract the names of each individual from the unformatted text string shown above and store them in a vector of some sort. When complete, your vector should contain the following entries:

```python
pattern = re.compile(r'[A-Za-z]+[\.\,\sA-Za-z]+')#question1
namelist = re.findall(pattern, text)
name = []
for i in namelist:
    name.append('\"'+i+'\"')
for i in name:
    print(i)
```

### 2.a Use your regex skills to rearrange the vector so that all elements conform to the standard“firstname lastname”, preserving any titles (e.g., “Rev.”, “Dr.”, etc) or middle/second names.

```python
stname = [] #question2(a)
for i in name:
    
    pattern = re.compile(r'[A-Za-z]+\,')
    find_name = re.findall(pattern, i)
    if len(find_name)>0:
        stname.append('"'+i.split(',')[-1][1:-1]+' '+find_name[0][0:-1]+'"')
    else:
        stname.append(i)
for i in stname:
    print(i)
```

###  b. (4 Points) Using your regex skills, construct a logical vector indicating whether a character has a title (i.e.,

```python
for i in stname:#question2(b)，this part we feel puzzle if the 'F.' is a title.
    pattern = re.compile(r'[A-Za-z]{2,}\.')
    a=re.findall(pattern, i)
    if ''.join(a) != '':
        print(i+'have a title :'+''.join(a))
```

### c. (4 Points) Using your regex skills, construct a logical vector indicating whether a character has a middle/second name.

```python
for i in stname:#question2(c)
    pattern = re.compile(r'\s[A-Za-z]+\s')
    a = re.findall(pattern, i)
    b = ''.join(a)
    if b != '': 
        sepattern = re.compile(r'\s[A-Za-z]+\s')
        print(b+' have a second name: '+''.join(re.findall(sepattern, b)))#extract the second name data
```

### 3. (6 Points) Consider the HTML string <title>+++BREAKING NEWS+++<title>. We would like to extract the first HTML tag (i.e., “<title>”). To do so we write the regular expression “<.+>”. Explain why this fails and correct the expression.

```python
htmltext = '<title>+++BREAKING NEWS+++<title>'#question3
pattern = re.compile(r'^<\w+>')
'''
when pattern = re.compile(r'<.+>'), '.' means you wnat to match any character including '<' and '>'.
Meanwhile, findall()use the greedy algorithm. So when the program match the '>'and '<' in the HTML string,
the program use the '.' to adsorb these two marks rather than think it is the end mark '>' or begin mark '<'.
The correct regex like the above.
'''
a = ''.join(re.findall(pattern, htmltext))
print(a)
```

## Part 2: Analyzing Chess Tournament Results

For Part 2 of Project 1, you’re given a text file (“Project1.txt”) with chess tournament results where the information has some structure. Your job is to create a Jupyter Notebook that generates a .CSV file with the following information for each of the chess players:
Player’s Name, Player’s State, Total Number of Points, Player’s Pre-Rating, and Average Pre Tournament Chess Rating of Opponents For the first player shown in the file excerpt below, that information would be:
Gary Hua, ON, 6.0, 1794, 1605
His “Average Pre Tournament Chess Rating of Opponents” score of 1605 was calculated by using the pre-tournament opponents’ ratings of 1436, 1563, 1600, 1610, 1649, 1663, 1716, and dividing by the total number of games played.
For each player we are provided with the total points they’ve won during the tournament and details on the results of their seven rounds of play. For each round we are given the unique ID of their opponent (an integer value) and an indicator of whether they won (‘W’), lost (‘L’), achieved a draw (‘D’), had a bye for that round (‘B’), or were unable to compete (‘U’).
If you have questions about the meaning of the remainder of the data or the results, please post them in the weekly Discussion Forum. Data science, like chess, is a game of back and forth...
The chess rating system (invented by a Minnesota statistician named Arpad Elo) has been used in many other contexts, including assessing the relative strength of employment candidates by human resource departments.

```python
with open('Project1.txt') as csvfile:
    csvdata = csv.reader(csvfile)
    for row in csvdata:
        print(row)
```

```python
'''
My idea is to extract the first row for two-row of each player and split them into two lists.
Because they have the same index in these two lists for the same player so that it can be easy to the operative.
'''
extractdata1 = []#including name, total pts, each round and opponent
extractdata2 = []#including state, Player’s Pre-Rating, each round
with open('Project1.txt') as csvfile:
    csvdata = csv.reader(csvfile)
    pattern = re.compile(r'[0-9]+\s\|\s[A-Za-z\-\s]+[\|A-Z0-9\.\s]*')#get data for extractdata1
    pattern2 = re.compile(r'[A-Z]+\s\|\s[0-9]+[\s\/A-Z\:\-\>\|0-9]*')#get data for extractdata2
    for row in csvdata:
        rowstring = str(row)
        #extractdata1
        a = ''.join(re.findall(pattern, rowstring))#remove other bother symbol
        if a != '':
            nre = []
            b = ''.join(re.sub('\|', ' ', a))#remove '|'
            cc = ''.join(re.findall('^[0-9]*', b)) #save the begin data
            c = ''.join(re.split('^[0-9]*', b))#remove the begine number
            d = ''.join(re.sub('\s+', ' ', c)) #let more than 1 whitespace become 1 whitespace
            #e = ''.join(re.findall('[0-9]\.[0-9A-Z\s]*', d)).strip() # get the data except name
            e = ''.join(re.findall('\s\s\s[\|A-Z0-9\.\s]*', a)).strip() # get the data except name
            ff = re.split('\|', e)[1:-1]#get ecah row data
            f = ''.join(re.split('[0-9]\.[0-9A-Z\s]*', d))#only get the data name
            nre.append(cc.strip())
            nre.append(f.strip())
            nre.extend(ff)
            extractdata1.append(nre)
            
            
            
        #extractdata2
        aa = ''.join(re.findall(pattern2, rowstring))
        if aa != '':
            nre = []
            bb = ''.join(re.findall(r'[A-Z]{2}\s', aa))# get the state data
            cc = ''.join(re.findall(r'[0-9]+[\s\/A-Z\:\-\>\|0-9]*', aa))#remove the state data
            dd = ''.join(re.findall(r'[0-9]+\s\/[\sA-Z\:\-\>0-9]*', aa))#find the data below name
            ee = ''.join(re.sub(r'[0-9]+\s\/[\sA-Z\:\-\>0-9]*',' ', cc))
            ff = ''.join(re.sub('\s+', ' ', ee)) #let more than 1 whitespace become 1 whitespace
            nre.append(bb.strip())
            nre.append(dd)
            nre.extend(ff.split('|')[1:-1])
            extractdata2.append(nre)
            
'''
extractdata1:
num,name,totalpit,rf1,rf2,rf3,rf4,rf5,rf6,rf7   (index,10)


extractdata1:
state,totalpit,pre-tournamentopponents,rf1,rf2,rf3,rf4,rf5,rf6,rf7  (index,10)

'''    
#print(extractdata1)
#print(extractdata2)
#get each player's pre-tournament rating
each_player_pre_tournament_rating = []#64
for i in extractdata2:
    a = ''.join(i)#change each row to string
    pattern = re.compile(r'\s[0-9]{3,4}')
    b = ''.join(re.findall(pattern, a)[:1]).strip()#get the each player's pre-tournament rating
    each_player_pre_tournament_rating.append(b)
#print(each_player_pre_tournament_rating)

#get the match result from extractdata1 and calculate the average pre-tournament rating for each player’s opponents
average_pre_tournament_rating_for_each_players_opponents = []
for i in extractdata1:
    j=3
    each_players_opponents = 0 
    opponum = 0 #calculate the number of opponent
    while j<len(i):
        #print(i[1])
        pattern = re.compile(r'[0-9]+')
        opponentnumlist = ''.join(re.findall(pattern, i[j]))#find the opponent of each round
        if opponentnumlist=='':
            j+=1
            continue
        opponentnum = int(opponentnumlist)
        each_players_opponents += int(each_player_pre_tournament_rating[opponentnum-1])#add the score of each opponent
        opponum+=1#opponent num add 1
        #print(each_player_pre_tournament_rating[opponentnum-1])
        j+=1
    average_pre_tournament_rating_for_each_players_opponents.append(round(each_players_opponents/opponum, 2))#get the average and round to 2 digit
#print(average_pre_tournament_rating_for_each_players_opponents)
```

```python
#finall write name(playername),total point(totalpoint), state(state), pre-tournament rating(), 
#average pre-tournament rating for each player’s opponents(average_pre_tournament_rating_for_each_peach_player_pre_tournament_ratinglayers_opponents). into
#pandas DataFrame and then .CSV file
playername = []
totalpoint = []
state = []
for i in extractdata1:
    playername.append(i[1].strip())
    totalpoint.append(i[2].strip())
for i in extractdata2:
    state.append(i[0].strip())
print(len(state))
```

```python
#start use pandas to create csv
pddata = []
pddata.append(playername)
pddata.append(totalpoint)
pddata.append(state)
pddata.append(each_player_pre_tournament_rating)
pddata.append(average_pre_tournament_rating_for_each_players_opponents)
#print(pddata)
columns = ['playername','totalpoint', 'state', 'each_player_pre_tournament_rating', 'average_pre_tournament_rating_for_each_players_opponents']
df = pd.DataFrame(pddata, columns)
df.to_csv('playerdata.csv')
```
