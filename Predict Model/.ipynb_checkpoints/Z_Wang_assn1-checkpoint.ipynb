### b) 
In Python (using, ex., numpy and matplotlib) generate samples of y according to the distribution given above using mu=1. Next, treat these samples as a data set for which you don't yet know  the true value of mu (pretend you don't know that mu=1) and must estimate it. Calculate SSE for a range of candidate values of mu (ex., mu_candidates=np.arange(0, 2.01, .1)). Plot SSE vs mu_candidates. Identify, on the plot, the value of mu that minimizes SSE.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
e = np.random.normal(0,1,2048)
```

```python
x = [i for i in range(2048)]
y = 1+e
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x,y)
plt.show()
```

```python
mu_candidates=np.arange(0, 2.01, .1)
```

```python
sse_list = []
for i in mu_candidates:
    sse = 0
    for j in y:
        sse += math.pow(j-i,2)
    sse_list.append(sse)
sse_list
```

```python
plt.figure(figsize=(15,10))
plt.scatter(mu_candidates,sse_list)
plt.show()
```

### So, we can see that the value of mu=1.00 that minimizes SSE

#### 3.3) 
b.Dividethe 20 households into three groups, placing the first seven households in the three group, the next six in the second group, and the last seven in the third group. Calculate the mean Y and mean X in each of the three groups. Transfer these means to the scatterplot and connect them with a simple nonparametric regression line.

```python
first = df1.loc[0:6,:]
second = df1.loc[7:13,:]
third = df1.loc[13:,:]
f_xm = first['consumers'].mean()
f_ym = first['acres'].mean()
s_xm = second['consumers'].mean()
s_ym = second['acres'].mean() 
t_xm = third['consumers'].mean()
t_ym = third['acres'].mean()
```

```python
x_mean = []
y_mean = []
x_mean.append(f_xm)
x_mean.append(s_xm)
x_mean.append(t_xm)
y_mean.append(f_ym)
y_mean.append(s_ym)
y_mean.append(t_ym)
```

```python
first['acres'].mean()
```

```python
second['acres'].mean()
```

```python
third['acres'].mean()
```

```python
#create nonparametric regression line
x = x_mean#new function
y = y_mean
xx_mean = np.mean(x)
yy_mean = np.mean(y)
xy_mean = (np.array(x_mean)*np.array(y_mean)).mean()
xx2_mean = (np.array(x_mean)*np.array(x_mean)).mean()
```

```python
a = (xx_mean*yy_mean-xy_mean)/(np.sqrt(xx_mean)-xx2_mean)
b = yy_mean-a*xx_mean
xx = np.array((df1['consumers']))
```

```python
a
```

```python
b
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x_mean,y_mean)
plt.plot(xx, a*xx+b)
plt.show()
```

Yes, it does. This regression line can help us to know the relationship between Acres/Gardener (Y ) and Consumers/Gardener (X) is positive relationship.
