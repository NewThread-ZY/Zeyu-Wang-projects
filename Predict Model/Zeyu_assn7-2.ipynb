```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def log_likelihood1(x,y,B):
    return ((np.log(np.exp(B*x)/(np.exp(B*x)+1))**y)).sum()+(np.log((1/(1+np.exp(B*x)))**(1-y))).sum()
```
