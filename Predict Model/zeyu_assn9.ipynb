## a

```python
theta1 = 0.6
thita1 = 0.4
x = [i for i in range(9)]
p1 = ((1+theta1*thita1)*(theta1+thita1))/(1+(thita1**2)+(2*theta1*thita1))
ps = [1]
ps.append(p1)
for i in range(7):
    ps.append(theta1*ps[-1])
plt.stem(x, ps, use_line_collection = True)
plt.ylim(0, 1)
```

```python
xx = [i for i in range(10000)]
np.random.seed(11)
n = 10000
eps = [np.random.normal()]
vt = [np.random.normal()]
for _ in range(1, n):
    vt_new = np.random.normal()
    eps.append(theta1*eps[-1]+vt_new+thita1*vt[-1])
    vt.append(vt_new)
x = np.random.normal(size=(n,))
beta = .1
y = beta*x + eps
plt.scatter(xx, eps)
```

```python
plt.scatter(xx, y)
```

```python
np.shape(x)
```

```python
x = x[:,None]
y = y
B = (np.linalg.inv(x.T@x)@(x.T@y))[0]
yhat = B*x
e = (y-yhat).flatten()
np.corrcoef(e[:-1],e[1:])[0,1]
```
