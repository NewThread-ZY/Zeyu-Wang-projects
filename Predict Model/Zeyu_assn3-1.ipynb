### Exercise D5.2

```python
df = pd.read_csv('https://socialsciences.mcmaster.ca/jfox/Books/Applied-Regression-3E/datasets/Sahlins.txt',
                      sep = '\s', names = ['','consumers','acres'], index_col = 0, skiprows = 1)

df
```

```python
x = df['consumers']
y = df['acres']
plt.plot(x,y,'o')
plt.xlabel("Consumers/Gardner")
plt.ylabel("Acres/Gardner");
```

```python
x = df['consumers']
y = df['acres']
ym = y-y.mean()
xm = x-x.mean()
B = (xm*ym).sum() / (xm**2).sum()
A = y.mean() - B * x.mean()
print (A, B)
```

```python
plt.plot(x,y,'o')
c = plt.axis()
plt.plot([c[0],c[1]], [A+B*c[0], A+B*c[1]], '--');
plt.xlabel('consumers')
plt.ylabel('acres')
plt.show();
```

Because we only use regressing Acres/Gardener on Consumers/Gardener. But in a society characterized by primitive communism, the product would be redistributed according to need, while each household would work in proportion to its capacity. And more acres means more work proportion. Since the work is a relationship with its capacity rathe than consumers. So in the y = a+bx+e model, b will be 0 while y means work and x mean consumers. In contrast, in a market society, work(acres) will be influenced by consumers. And its slop is b. And because they have a positive correlation relationship so that the slop b is positive.

```python
#calculate Se
ee = 0
xx = list(x)
yy = list(y)
n = len(y)
for i in range(n):
    ee += (yy[i]-(A+B*xx[i]))**2
se2 = ee/(n-2)
se = se2**0.5
se
```

```python
yhat = A + B*x
eps = y - yhat
TSS = ( (y-y.mean())**2 ).sum()
RSS = ( (y-yhat)**2 ).sum()
RegSS = TSS - RSS
R2 = RegSS / TSS
r = R2**0.5
r
```

```python
print(f'A={A} B={B} Se = {se} r2 = {R2} r = {r}')
```

### Exercise D6.2

```python
#get variance of e
x = df['consumers']
y = df['acres']
x
```

```python
ee = []
xx = list(x)
yy = list(y)
n = len(y)
for i in range(n):
    ee.append(yy[i]-(A+B*xx[i]))
ee = np.array(ee)
ee_var = (ee*ee).sum()/(n-2)
ee_var
```

```python
print(n)
```

```python
var_A = (ee_var*(x*x).sum())/(n*(((x-x.mean())**2).sum()))
var_B = ee_var/((x-x.mean())**2).sum()
```

```python
standard_errors_A = var_A**0.5
standard_errors_B = var_B**0.5
```

```python
print(f'standard errors of the least-squares intercept={standard_errors_A}')
print(f'standard errors of the least-squares slope.={standard_errors_B}')
```

For 95% confidence intervals for A and B

```python
AAplus = A+2.101*standard_errors_A
AAmin = A-2.101*standard_errors_A
print(f'{AAmin}<A<{AAplus}')
```

```python
BBplus = B+2.086*standard_errors_B
BBmin = B-2.086*standard_errors_B
print(f'{BBmin}<B<{BBplus}')
```

So we have 95% confidence to conclude slop is greater than 0 but intercept isn't always greater than 0.
