```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 21)
```

# Problem 1: Hyperbolic Tangent

```python
y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
plt.plot(x, y)
```

#### Derivate of Hyperbolic Tangent

```python
y_hat = 1-((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))**2
```

```python
plt.plot(x, y_hat)
```
