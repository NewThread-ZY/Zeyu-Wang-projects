```python
import pandas as pd
```

```python
import tensorflow as tf
```

```python
h_data = pd.read_csv('https://gist.githubusercontent.com/keeganhines/59974f1ebef97bbaa44fb19143f90bad/raw/d9bcf657f97201394a59fffd801c44347eb7e28d/Hitters.csv')
```

```python
h_data.head()
```

# 2.Clean data

- 1. I would remove all of the Nan data
- 2. I will use one hot code to encode the category data
- 3. I will normalization the numerical data

```python
len(h_data)
```

```python
data = h_data.dropna()
data = data.reset_index(drop = True)
data.head()
```

```python
data.iloc[0,:]
```

```python
#one hot encoding League
onehot_League = pd.get_dummies(data['League'],prefix='League')
data = data.drop('League',axis = 1)
data = data.join(onehot_League)
```

```python
#one hot encoding Division
onehot_Division = pd.get_dummies(data['Division'],prefix='Division')
data = data.drop('Division',axis = 1)
data = data.join(onehot_Division)
```

```python
#one hot NewLeague
onehot_NewLeague = pd.get_dummies(data['NewLeague'],prefix='NewLeague')
data = data.drop('NewLeague',axis = 1)
data = data.join(onehot_NewLeague)
```

```python
data.head()
```

```python
# Let the salary become the last columns
Salary = data.Salary
data = data.drop('Salary',axis = 1)
data = data.join(Salary)
```

```python
data.head()
```

```python
# Drop the name
data = data.drop('Unnamed: 0',axis = 1)
```

```python
data.head()
```

```python
#keep data for recovrying the salary
salary_min = data.Salary.min()
salary_max = data.Salary.max()
```

```python
#Normalize the data
data = (data-data.min())/(data.max()-data.min())
data.head()
```

# 3.Split data into train and test

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
tf.random.set_seed(42)
```

```python
np.random.seed(42)
model = Sequential()
```

```python
#Get how many feature the dataset has
input_dim = len(X.columns)
```

```python
model.add(Dense(22, activation='relu', input_dim=input_dim))
model.add(Dense(6,  activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='sgd', loss='mse', \
metrics=['mse'])
model.summary()
```

```python
history = model.fit(X, y, epochs=100, batch_size=5, \
verbose=1, validation_split=0.2, \
shuffle=False)
```

```python
model.metrics_names
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
y_predicted = model.predict(X_test)
```

```python
# print the predicted classes
print("Predicted probability for each of the examples belonging to class 1: ")
print(y_predicted)
```

```python
y_test
```

```python
model.evaluate(X, y, batch_size=None, verbose=0)
```
